# Kwantyzacja kolorw



## Wstp
Standardowy kolorowy obraz rozmiarw $n \times m$ jest reprezentowany przez $n \cdot m \cdot 3$ liczb cakowitych z przedziau $[0, 255]$. Daje to cznie ponad 16 milionw moliwych kolorw. Kwantyzacja kolorw polega na drastycznym zredukowaniu liczby uywanych kolorw w celu redukcji rozmiaru pliku, przy jednoczesnym jak najlepszym zachowaniu jakoci obrazu.
Najprostsz metodkwantyzacji jest uycie metody $k$-means. Dziaa znakomicie jeli chodzi o redukcj kolorw, jednak jej natura wpywa te na ostateczny wygld obrazka:





Metoda ([$k$-means](https://en.wikipedia.org/wiki/K-means_clustering)) kwantyzuje obraz minimalizujc bd redniokwadratowy ([mean squared error - MSE](https://en.wikipedia.org/wiki/Mean_squared_error)) pomidzy oryginalnym a zrekonstruowanym obrazkiem. 
Jednake zupenie nie obchodzi jej jakich kolorw ostatecznie uyje. Okazuje si jednak, e jeli narzucimy koszty uycia kolorw to bdziemy w stanie wpyn na jako kwantyzacji. Wtedy jednak metoda $k$-means nie bdzie ju optymalna.

Zmiana metody kwantyzacji moe wic wpyn na ostateczn jako skwantyzowanego obrazu. I o tym wanie jest ponisze zadanie.


## Zadanie

Celem zadania jest napisanie algorytmu klasteryzacji, ktry minimalizuje inn funkcj celu ni $k$-means.
Na pocztek zdefiniujmy sobie *koszt koloru*. Kolor RGB rozumiemy jako wektor trjwymiarowy ze zbioru $[0,1,...,255]^3$, a koszt koloru $c$ zdefiniujemy jako
odlego euklidesow od najbliszego "prostego" koloru w przestrzeni RGB, gdzie prostymi kolorami nazywamy: czarny $(0,0,0)$, biay $(255,255,255)$, czerwony $(255,0,0)$, zielony $(0,255,0)$, niebieski $(0,0,255)$, ty $(255,255,0)$, magenta $(255,0,255)$ oraz cyjan $(0,255,255)$. Szczegy znajdziesz w funkcji `color_cost`.
Z tak zdefiniowan funkcj kosztu najbardziej kosztowny kolor to $(127,127,127)$, czyli szary. 
Interpretacja tej funkcji kosztu jest wic taka, e bdziemy chcieli wymusi na algorytmie kwantyzacji, aby preferowa kolory, ktre s bardziej jaskrawe i ywe.

Nasz funkcj celu bdzie waona suma poszczeglnych skadnikw:
1. `MSE`: bd redniokwadratowy pomidzy oryginalnym a zrekonstruowanym obrazkiem --- ten skadnik odpowiada za to, eby skwantyzowany obraz by bliski oryginaowi.
2. `max_color_cost`: najwikszy z kosztw uytych kolorw --- ten skadnik wzmacnia jaskrawo kolorw, poprzez kontrol najmniej jaskrawego koloru.
3. `mean_color_cost`: redni koszt koloru liczony po wszystkich pikselach --- ten skadnik wymusza, aby uyte kolory byy rednio odpowiednio jaskrawe.

Ostatecznym kryterium jakoci kwantyzacji bdzie $$2\cdot MSE  + 21\cdot max\_color\_cost + 42\cdot mean\_color\_cost.$$

Powysze kryterium i wszystkie funkcje, o ktrych powyej jest mowa, s zaimplementowane poniej przez nas. 

W zadaniu zakadamy, e liczba kolorw jest z gry ustalona i wynosi $37$. Twj algorytm ma wic skwantyzowa obraz na $37$ kolorw w taki sposb, aby zminimalizowa powysz funkcj celu.

```python
######################### NIE ZMIENIAJ TEJ KOMRKI ##########################
import os
import numpy as np
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import typing

import glob
```

## adowanie danych

```python
######################### NIE ZMIENIAJ TEJ KOMRKI ##########################

# Klasa, ktra uatwia wczytywanie obrazkw z danego folderu
class ImageDataset:
    def __init__(self, image_dir: str):
        self.filelist = glob.glob(image_dir + "/*.jpg")
        self.IMAGE_DIMS = (512, 512)

    def __len__(self):
        return len(self.filelist)

    def __getitem__(self, idx) -> np.ndarray:
        with Image.open(self.filelist[idx]) as image:
            image = image.convert('RGB')
            image = image.resize(self.IMAGE_DIMS)
            return np.array(image)

    def __iter__(self) -> typing.Iterator[np.ndarray]:
        return (self[i] for i in range(len(self.filelist)))
```

## Kod z kryterium oceniajcym

```python
######################### NIE ZMIENIAJ TEJ KOMRKI ##########################

# Poniej znajdziesz definicje MSE oraz kosztu uycia kolorw
# Pamitaj, eby przy ewaluacji liczy je w przestrzeni RGB, tzn. na wartociach cakowitych z przdziau [0, 255]
# Skalowanie jest dopuszczalne tylko podczas treningu!

# Zdefiniujmy kryterium oceny jakoci kwantyzacji
# Uyjemy do tego bdu redniokwadratowego (mean square error - MSE)
def mse(img, img_quant):
  return ((img_quant.astype(np.float32) - img.astype(np.float32))**2).mean()


# Nastpnie zdefinujmy koszt uycia kolorw
# Im bliej danemu kolorowi do "prostych" kolorw, tym mniejszy koszt jego uycia
def color_cost(img_quant):
    vertices = np.array([
        [0, 0, 0], [0, 0, 255], [0, 255, 0], [0, 255, 255],
        [255, 0, 0], [255, 0, 255], [255, 255, 0], [255, 255, 255]
    ])
    
    pixels = img_quant.reshape(-1,3)
    
    differences = pixels[:, np.newaxis, :] - vertices[np.newaxis, :, :]
    squared_distances = np.sum(differences**2, axis=2)
    costs = np.sqrt(np.min(squared_distances, axis=1))

    return np.mean(costs), np.max(costs)


# Cakowite kryterium zdefiniowane w treci zadania
def quantization_score(img, img_quant):
    assert img.dtype == np.uint8
    assert img_quant.dtype == np.uint8
    assert len(np.unique(img_quant.reshape(-1,3), axis=0)) == 37
       
    mse_cost = mse(img, img_quant)
    mean_color_cost, max_color_cost = color_cost(img_quant)
    score = mse_cost * 2 + max_color_cost * 21 + mean_color_cost * 42
    print(f'MSE: {mse_cost:.4f}, max_color_cost: {max_color_cost:.4f}, mean_color_cost: {mean_color_cost:.4f}')
    print(f'Score: {score:.4f}')
    return score
```

# Twoje rozwizanie

Ta sekcja to miejsce na Twoje rozwizanie. Tylko tutaj powiniene wprowadza zmiany!

```python
"""
Opis rozwizania:
Sie przetwarza obraz za pomoc warstw konwolucyjnych i zwraca 37 kolorw po 3 kanay (111 neuronw na wyjciu).
Trening sieci odbywa si za pomoc bezporedniej minimalizacji straty zdefiniowanej wg warunkw punktacji
(wprost utworzony zostaje obraz wg kolorw zwrconych przez sie a nastpnie obliczany jest koszt kwantyzacji)
Dla kadego obrazu tworzona jest nowa instancja modelu, by osign jak najleszy wynik kwantyzacji

Przy tworzeniu skwantyzowanego obrazu funkcja unique_centroids dba o to, by uyte zostao dokadnie 37 rnych kolorw. 
Jej wynik nie jest uwzgldniany przez funkcj straty, co jednak jest rekompensowane przez zapamitywanie ostatniego najlepszego obrazu.
"""

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
import numpy as np

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

class ConvNet(nn.Module):
    def __init__(self):
        super(ConvNet, self).__init__()
        
        self.conv1 = nn.Conv2d(in_channels=3, out_channels=32, kernel_size=3, stride=1, padding=1)
        self.conv2 = nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3, stride=1, padding=1)
        self.conv3 = nn.Conv2d(in_channels=64, out_channels=128, kernel_size=3, stride=1, padding=1)
        
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2, padding=0)
        
        self.fc1 = nn.Linear(128 * 64 * 64, 512)
        self.fc2 = nn.Linear(512, 256)
        self.fc3 = nn.Linear(256, 111)
        
    def forward(self, x):
        # Zamiana kolejnoci wymiarw z (N, H, W, C) na (N, C, H, W)
        x = x.permute(0, 3, 1, 2).to(device)

        # Normalizacja pikseli do zakresu [0, 1]
        x = x / 255.0
        
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = self.pool(F.relu(self.conv3(x)))
        
        x = x.reshape(-1, 128 * 64 * 64)
        
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = torch.sigmoid(self.fc3(x))
        
        return x


def quantize_image(images, centroids):
    N, H, W, C = images.shape
    _, K, _ = centroids.shape

    images_flat = images.view(N, -1, C)
    
    distances = torch.cdist(images_flat, centroids)
    nearest_centroids = torch.argmin(distances, dim=2)
    
    quantized_images_flat = centroids.gather(1, nearest_centroids.unsqueeze(-1).expand(-1, -1, C))
    quantized_images = quantized_images_flat.view(N, H, W, C)
    
    quantized_images_flat = quantized_images.view(N, -1, C)
    for i in range(N):
        unique_colors = torch.unique(quantized_images_flat[i], dim=0)
        missing_colors = set(range(K)) - set(nearest_centroids[i].unique().tolist())
        
        if missing_colors:
            missing_colors = list(missing_colors)
            idx = torch.randperm(images_flat.shape[1])[:len(missing_colors)]
            quantized_images_flat[i, idx] = centroids[0, missing_colors]
    
    quantized_images = quantized_images_flat.view(N, H, W, C)

    return quantized_images


def my_tensor_mse(img, img_quant):
    return ((img_quant - img) ** 2).mean()


def my_color_cost(img_quant):
    vertices = torch.tensor([
        [0, 0, 0], [0, 0, 255], [0, 255, 0], [0, 255, 255],
        [255, 0, 0], [255, 0, 255], [255, 255, 0], [255, 255, 255]
    ], dtype=torch.float32, device=device)

    pixels = img_quant.view(-1, 3).float()
    
    differences = pixels[:, None, :] - vertices[None, :, :]
    squared_distances = torch.sum(differences ** 2, dim=2)
    
    costs = torch.sqrt(torch.min(squared_distances, dim=1)[0])
    
    return torch.mean(costs), torch.max(costs)


def loss(old_image, new_image):
    old_image = (old_image * 255).float().to(device)
    new_image = (new_image * 255).float().to(device)

    mse = my_tensor_mse(old_image, new_image)
    
    mean_color_cost, max_color_cost = my_color_cost(new_image)
  
    loss_value = 2 * mse + 21 * max_color_cost + 42 * mean_color_cost
    
    return loss_value


def unique_centroids(centroids):
    cloned1 = centroids.clone()
    cloned2 = (centroids.clone() * 255).to(torch.uint8)
    b_size = centroids.shape[0]

    unique_set = set()

    for i in range(b_size):
        for j in range(37):
            color_as_tuple = cloned2[i][j].tolist()
            direction = 1
            while tuple(color_as_tuple) in unique_set:
                if color_as_tuple[0] + direction > 255:
                    direction *= -1
                color_as_tuple[0] += direction
                cloned1[i][j][0] += direction / 255

            unique_set.add(tuple(color_as_tuple))
            cloned2[i][j][0] = color_as_tuple[0]

    return cloned1


def your_quantization_algorithm(img, n_clusters=37, num_epochs=100, learning_rate=0.0001):
    img_with_batch = np.expand_dims(img, axis=0)  # Shape: (1, 512, 512, 3)

    batch_size = 1
    batch_img = np.repeat(img_with_batch, batch_size, axis=0)  # Shape: (16, 512, 512, 3)

    img_tensor = torch.tensor(batch_img, dtype=torch.float32).to(device) / 255  # Shape: (16, 512, 512, 3)

    model = ConvNet().to(device)
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)

    last_img = None
    last_loss = float("inf")
    for epoch in range(num_epochs):
        model.train()

        centroids = model(img_tensor)
        centroids = centroids.view(batch_size, n_clusters, 3)

        quantized_img = quantize_image(img_tensor, unique_centroids(centroids))

        loss_value = loss(img_tensor, quantized_img)
        
        optimizer.zero_grad()

        loss_value.backward()
        optimizer.step()

        img_as_numpy = (quantized_img.squeeze().detach().cpu().numpy() * 255).astype(np.uint8)
        mse_cost = mse(img, img_as_numpy)
        mean_color_cost, max_color_cost = color_cost(img_as_numpy)
        score = mse_cost * 2 + max_color_cost * 21 + mean_color_cost * 42

        if last_loss > score:
            last_loss = score
            last_img = img_as_numpy

    return last_img
```

```python
# Wywietl zestawienie oryginau i zkwantyzowanego obrazka
def show_quantization_results(original, quantized):
        _, ax = plt.subplots(1, 2, figsize=(8, 4))
        ax[0].imshow(original)
        ax[0].set_title("Orygina")
        ax[0].axis("off")

        ax[1].imshow(quantized)
        ax[1].set_title("Wynik kwantyzacji")
        ax[1].axis("off")
        plt.show()
```

```python
def evaluate_algorithm(quantization_algorithm, data_dir):
    dataset = ImageDataset(data_dir)
    scores = []
    for image in dataset:
        quantized_image = quantization_algorithm(image)
        show_quantization_results(image, quantized_image)
        score = quantization_score(image, quantized_image)
        scores.append(score)
    return np.mean(scores)
```
